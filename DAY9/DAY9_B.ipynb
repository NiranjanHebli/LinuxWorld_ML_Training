```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
state=dataset["State"]
```

```python
state
```

```python
dummy_states=pd.get_dummies(state)
```

```python
dummy_states
```

```python
type(dummy_states)
```

```python
dummy_states.iloc[:,0:2]
```

```python
final_dummy_state=dummy_states.iloc[:,0:2]
```

```python
final_dummy_state
```

```python

  
result = pd.concat([final_dummy_state,dataset ], axis=1,join='inner')
```

```python
result
```

```python
final_result=result.drop(['State','Profit'], axis=1)
```

```python
y=dataset["Profit"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_result, y,test_size=0.33, random_state=42)
```

```python
model=LinearRegression()
```

```python
mymodel=model.fit(X_train,y_train)
```

```python
mymodel.predict(X_test)
```

```python
y_test
```

```python
y_pred=mymodel.predict(X_test)
```

```python
from sklearn import metrics 
```

```python
metrics.mean_absolute_error(y_test,y_pred)
```
